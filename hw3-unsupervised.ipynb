**Задание 1.** (1 балл)

Сделайте кластеризацию данных методом `KMeans` (собственная имплементация и готовая реализация), обучив модель на данных со сниженной за счет PCA размерностью. В данном случае мы подскажем, что нужно искать именно 6 кластеров, но в общем случае мы не будем знать, сколько кластеров надо искать.

Параметры:

- **n_clusters** = n_classes (число уникальных меток целевого класса)
- **n_init** = 100
- **random_state** = RANDOM_STATE (для воспроизводимости результата)

Остальные параметры со значениями по умолчанию.

````python
# Ваш код здесь
def compute_inertia_fast(X, centers, labels):
    inertia = 0
    unique_labels = np.unique(labels)
    for i in unique_labels:
        mask = labels == i
        if np.any(mask):
            inertia += np.sum((X[mask] - centers[i]) ** 2)
    return inertia

class CustomKMeans:
    def __init__(self, n_clusters=8, n_init=100, max_iter=300, random_state=None):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.max_iter = max_iter
        self.random_state = random_state
        self.cluster_centers_ = None
        self.labels_ = None
        self.inertia_ = float('inf')

    def fit_predict(self, X):
        if self.random_state is not None:
            np.random.seed(self.random_state)

        n_samples = X.shape[0]
        best_centers = None
        best_labels = None

        X_squared = np.sum(X ** 2, axis=1, keepdims=True)

        for init in range(self.n_init):
            idx = np.random.choice(n_samples, self.n_clusters, replace=False)
            centers = X[idx].copy()

            for iteration in range(self.max_iter):
                centers_squared = np.sum(centers ** 2, axis=1)
                distances = X_squared - 2 * X.dot(centers.T) + centers_squared

                labels = np.argmin(distances, axis=1)

                new_centers = centers.copy()
                for i in range(self.n_clusters):
                    mask = labels == i
                    if np.any(mask):
                        new_centers[i] = X[mask].mean(axis=0)

                if np.allclose(centers, new_centers):
                    centers = new_centers
                    break

                centers = new_centers

            inertia = compute_inertia_fast(X, centers, labels)

            if inertia < self.inertia_:
                self.inertia_ = inertia
                best_centers = centers.copy()
                best_labels = labels.copy()

        self.cluster_centers_ = best_centers
        self.labels_ = best_labels
        return self.labels_

custom_kmeans = CustomKMeans(n_clusters=n_classes, n_init=100, max_iter=300, random_state=RANDOM_STATE)
custom_kmeans_labels = custom_kmeans.fit_predict(X_pca)

kmeans = KMeans(n_clusters=n_classes, n_init=100, random_state=RANDOM_STATE)
kmeans_labels = kmeans.fit_predict(X_pca)

print(f"Инерция собственной реализации KMeans: {custom_kmeans.inertia_:.2f}")
print(f"Инерция sklearn KMeans: {kmeans.inertia_:.2f}")
````
